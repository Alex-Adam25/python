```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def psi(x):
    return 2*A*np.cos(k0*x)*np.sin(dk*(x))/x
```

```python
A=1
k0=1
dk=1
x=np.linspace(-20,20,1000)
y=psi(x)
plt.plot(x,y)
plt.title(f"A(k0)={A}; k0={k0}; dk={dk}")
plt.xlabel("x")
plt.ylabel(r"$Re(\Psi(r))$")
plt.savefig("epu5/1.png",dpi=300)
```
