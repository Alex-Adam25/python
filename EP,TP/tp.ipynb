```python
import numpy as np
import matplotlib.pyplot as plt

a=100
v=1
x0=np.log(v**2)/(-a)
t=np.linspace(0,1,1000)


x=[2/a*np.log(np.cosh(v*t*a/2))+x0 for t in t]
x_inf=v*t
plt.plot(t,x)
plt.plot(t,x_inf)
```

```python
import scipy.integrate as integrate
import pandas as pd
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0.15,5,1000)
w=1
a=1
v0=0
x0=0.2
m=1
def v(x):
    return w**2*x**2/2 + a**2/(2*x**2)

e=0.5*m*v0**2+v(x0)

def f(x):
    return 1/np.sqrt(2*(e-v(x)))

xmin=minimize(lambda x: abs(v(x)-e),x0=0.00001).x+0.0001
xmax=minimize(lambda x: abs(v(x)-e),x0=10).x-0.0001

x_i=np.linspace(xmin,xmax,1000)

#x_full=np.concatenate((x_i, np.flip(x_i)), axis=0)
    


v_list=[w**2*x**2/2 + a**2/(2*x**2) for x in x]
plt.plot(x,v_list)
```

```python
I=np.array([integrate.quad(f,x0,x)[0] for x in x_i])


plt.plot(I,x_i)
```
