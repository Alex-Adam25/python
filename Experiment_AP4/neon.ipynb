```python
import matplotlib.pyplot as plt
import numpy as np
img = plt.imread("langzeit.png")
fig, ax = plt.subplots()
ax.imshow(img)
ax.axes.yaxis.set_visible(False)
xa=20
xe=870
x=np.linspace(20,870,9)
x5=x+(xe-xa)/16
x5=x5[:-1]

def rundennachkomma(a,n):
#a ist array welches gerundet werden soll
#n ist die anzahl der nachkommastellen
#Erzeugt array mit den selben dimensionen wie a nur als string array(max 20 char)
    ar=np.empty_like(a,dtype='<U20')
    for i,v in enumerate(a):
        if isinstance(v, str):
            ar[i]=v
        else:
            ar[i]="{:.{}f}".format(v,n)
       # print(ar)
    return ar
    
for i in x:
    ax.axvline(i,color='r',alpha=0.85)
for i in x5:
    ax.axvline(i,color='r',alpha=0.75,linestyle="dashed")
ax.set_xticks(x)
x_ticks_labels = ['0','10','20','30','40','50','60','70','80']
ax.set_xticklabels(x_ticks_labels, rotation='horizontal', fontsize=12)
plt.savefig("neonxlabel.png",dpi=300)
```
