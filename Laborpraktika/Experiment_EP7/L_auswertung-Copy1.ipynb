# NTC

```python
T_NTC,R_1=txt2arr("L_NTC")
R=R_1*10**3
R_err=20
T25=298.15
R25=10000
T0=273.15
B=3988

def ntc_func(T):
    T=T+T0
    return R25*np.exp(-B*(1/T25-1/T))
T_theo=np.linspace(min(T_NTC),max(T_NTC),10000)
R_theo=ntc_func(T_theo)
```

```python
plt.errorbar(x=T_NTC,y=R,yerr=R_err,fmt='.',c='b',capsize=5,ms=7,label="Messwerte")
plt.errorbar(x=T_theo,y=R_theo,fmt='-',c='k',capsize=2,ms=5,label="theoretische Kurve")
plt.ylabel("Widerstand R [$\Omega$]")
plt.xlabel("Temperatur T [째C]")
#plt.yscale("log")
draw("Widerstand des NTCs")
save("NTC")
```

```python
align='c'
head=["Temperatur [째C]","Widerstand [R]"]
val=[T_NTC,R]
err=[0.1,R_err] 
cap="Eichung des NTCs"
lab="ntc"
nk=[1,0]
print(table(align,head,val,err,cap,lab,nk,longtable=False,copy=False))
```
