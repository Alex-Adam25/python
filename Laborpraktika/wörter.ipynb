```python
ap1=np.array(["EF", "T3", "M15", "M3", "GO1", "GO2"])
ap2=np.array(["E2", "AP1", "AP2", "AP4", "AP9", "AS1"])
ep=np.array(["EP2", "EP3", "EP4", "EP6","EP7"])
alle=np.concatenate((ap1,ap2,ep),axis=None)
anzahl=len(alle)
```

```python
lea_ap1=[916, 1318, 2122, 1572, 1422, 2013]
lea_ap2=[1639, 2885, 2343, 2073, 2475, 2007]
lea_ep=[1905, 2037, 2486, 2797,2030]
lea_alle=np.concatenate((lea_ap1,lea_ap2,lea_ep),axis=None)
lea_sum=np.array([np.sum(lea_alle[0:i]) for i in range(anzahl)])
```

```python
alex_ap1=[781,1794,m15,1658,1326,1407]
alex_ap2=[2194,2528,2943,1601,1873,1572]
alex_ep=[2047,2086,1923,2709,2586]
alex_alle=np.concatenate((alex_ap1,alex_ap2,alex_ep),axis=None)
alex_sum=np.array([np.sum(alex_alle[0:i]) for i in range(anzahl)])
```

```python
felix_ap1=[660,1688,2557,1560,1703,3396]
felix_ap2=[2354,3759,2633,2227,2753,1592]
felix_ep=[2555,2509,2853,3477,3032]
felix_alle=np.concatenate((felix_ap1,felix_ap2,felix_ep),axis=None)
felix_sum=np.array([np.sum(felix_alle[0:i]) for i in range(anzahl)])
```

```python
x=np.arange(0,anzahl)
plt.plot(x,lea_alle,'--',label="Lea Jan")
plt.plot(x,alex_alle,'--',label="Alex Fredi")
plt.plot(x,felix_alle,'--',label="Felix Ali")

plt.ylabel("Wrter")
plt.xlabel("Protokoll")
ax = plt.gca()
ax.set_xticks(x)
ax.set_xticklabels(alle, rotation='horizontal', fontsize=12)
draw("Wrter des Protokoll",size=(16,9))
save("proProtokoll")
```

```python
x=np.arange(0,anzahl)
plt.plot(x,lea_sum,'--',label=f"Lea Ende:{lea_sum[-1]}")
plt.plot(x,alex_sum,'--',label=f"Alex Ende:{alex_sum[-1]}")
plt.plot(x,felix_sum,'--',label=f"Felix Ende:{felix_sum[-1]}")

plt.ylabel("Wrter")
plt.xlabel("Protokoll")
ax = plt.gca()
ax.set_xticks(x)
ax.set_xticklabels(alle, rotation='horizontal', fontsize=12)
draw("Summe der Wrter bis inkl diesem Protokoll",size=(16,9))
save("summe")
```

```python
x=np.arange(0,anzahl)
plt.plot(x,lea_alle-alex_alle,'--',label="Lea")

plt.ylabel("Wrter diff")
plt.xlabel("Protokoll")
ax = plt.gca()
ax.set_xticks(x)
ax.set_xticklabels(alle, rotation='horizontal', fontsize=12)
draw("Lea-Alex pro Protokoll; Pos: Lea mehr; Neg: Alex mehr",size=(16,9))
```
