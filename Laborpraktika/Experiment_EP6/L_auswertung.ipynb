```python
B=3500
R25=10e3
T25=298.15
def NTC_func(T):
    T=T+273.15
    return R25*np.exp(-B*(1/T25-1/T))

R0=1000
alpha=3.85e-3
def PTC_func(T):
    return R0*(1+alpha*T)

PTC_I_mess=np.array([9.19,8.92,12.52])*10**-3
PTC_U_mess=np.full(len(PTC_I_mess),10)
PTC_R_mess=PTC_U_mess/PTC_I_mess

NTC_I_mess=np.array([0.73,0.96,16*10**-3])*10**-3
NTC_U_mess=PTC_U_mess
NTC_R_mess=NTC_U_mess/NTC_I_mess

NTC_T_mess=PTC_T_mess=[20,30,-40]

#NTC_R_mess=np.array([339000.0,12780.0,12360.0])
#PTC_R_mess=np.array([812,1076,1122])
```

```python
T=np.linspace(-50,30,10000)
NTC_R=NTC_func(T)
plt.plot(T,NTC_R*10**-3,label="theoretische Kurve",color="black")
# plt.xscale("log")
#plt.yscale("log")
plt.xlabel("Temperatur [°C]")
plt.ylabel("Widerstand [k$\Omega$]")

#plt.scatter(NTC_T_mess,NTC_R_mess*1e-3,marker="x",s=5, linewidth = 20,color="red",label="Messwerte")
plt.errorbar(NTC_T_mess,NTC_R_mess*10**-3,yerr=[0.5,0.5,30],fmt='.',ms=10,mew=2,capsize=10,color="blue",label="Messwerte ")

draw("Widerstand des NTCs")
save("NTC")
```
