```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Projekte_RKI/Nowcasting.html?nn=13490888


path = 'Nowcasting_Zahlen_csv.csv'
data = pd.read_csv(path,encoding='latin1',delimiter=";")
data = data[:-18]
data['Datum des Erkrankungsbeginns']= pd.to_datetime(data['Datum des Erkrankungsbeginns'],format="%d.%m.%Y")

data=data.replace(".","0,0")

data['Punktschätzer der Reproduktionszahl R']=data['Punktschätzer der Reproduktionszahl R'].str.replace(',', '.').astype(float)
data['Untere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R']=data['Untere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R'].str.replace(',', '.').astype(float)
data['Obere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R']=data['Obere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R'].str.replace(',', '.').astype(float)
data['Punktschätzer des 7-Tage-R Wertes']=data['Punktschätzer des 7-Tage-R Wertes'].str.replace(',', '.').astype(float)
data['Untere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes']=data['Untere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes'].str.replace(',', '.').astype(float)
data['Obere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes']=data['Obere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes'].str.replace(',', '.').astype(float)



R_pkt=data['Punktschätzer der Reproduktionszahl R'].tolist()
R_7dy=data['Punktschätzer des 7-Tage-R Wertes'].tolist()

datum=data['Datum des Erkrankungsbeginns'].tolist()
#R_pkt=pd.Series(R,index=datum)

R_pkt_unten=R_pkt-data['Untere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R']
R_pkt_oben=data['Obere Grenze des 95%-Prädiktionsintervalls der Reproduktionszahl R']-R_pkt

R_7dy_unten=R_7dy-data['Untere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes']
R_7dy_oben=data['Obere Grenze des 95%-Prädiktionsintervalls des 7-Tage-R Wertes']-R_7dy


plt.errorbar(x=datum,y=R_pkt,capsize=3 ,yerr=np.array([R_pkt_unten.tolist(),R_pkt_oben.tolist()])) 
plt.gcf().set_size_inches(10, 5)
plt.gca().get_ylim()[1]
plt.ylim(0,plt.gca().get_ylim()[1])
plt.grid(b=True, which='major',  linewidth=1)
plt.grid(b=True, which='minor',  linewidth=0.2)
plt.minorticks_on()
plt.title("R Punkt ")
plt.show()

plt.errorbar(x=datum,y=R_7dy,capsize=3 ,yerr=np.array([R_7dy_unten.tolist(),R_7dy_oben.tolist()])) 
plt.gcf().set_size_inches(10, 5)
plt.gca().get_ylim()[1]
plt.ylim(0,plt.gca().get_ylim()[1])
plt.grid(b=True, which='major',  linewidth=1)
plt.grid(b=True, which='minor',  linewidth=0.2)
plt.minorticks_on()
plt.title("R 7 Tage avg")
plt.ylim(0.5,2)
plt.show()
```
