```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Testzahl.html

#data = pd.read_csv('testdaten.txt', sep=" ", header=None,thousands=".")
#data.columns = ["KW", "Anzahl Testungen", "Positiv getestet", "Positivenquote in %","Anzahl Ã¼bermittelnde Labore"]
#data['Positivenquote in %'] = data["Positivenquote in %"].str.replace(',', '').astype(float)/100

data=pd.read_excel("Testzahlen-gesamt.xlsx",sheet_name="Testzahlen",usecols="B:F",skiprows=2,skipfooter=1)
data=data[1:]

data["Kalenderwoche 2020"]=pd.to_numeric(data["Kalenderwoche 2020"])
fig, ax1 = plt.subplots()


ax1.plot(data["Kalenderwoche 2020"],data["Anzahl Testungen"],label="Anzahl Testungen")

fig.set_size_inches(10, 7)
ax1.get_ylim()[1]
ax1.ticklabel_format(style='plain')
ax1.set_ylim(0,ax1.get_ylim()[1])
ax2 = ax1.twinx()

ax2.plot(data["Kalenderwoche 2020"], data["Positiv getestet"],label="Positiv getestet",c='r')

plt.grid(b=True, which='major',  linewidth=1)
plt.grid(b=True, which='minor',  linewidth=0.2)
plt.minorticks_on()
plt.title("")
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2,loc='upper center')
plt.savefig("test.pdf")
plt.show()
```
