```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import os
```

```python
shape_df = gpd.read_file('VG250_LAN.shp') #Laden des Shapefiles im gleichnamigen ordner (muss so) und der entsprechenden zusätzlichen dateien
shape_df=shape_df.drop(shape_df.index[[33,34]]) #Entfernen von Baden-WÃ¼rttemberg (Bodensee) & Bayern (Bodensee
shape_df.head()
shape_df.rename(columns={'GEN':'Bundesland'}, inplace=True)
```

```python
def split(strng, sep, pos):
    strng = strng.split(sep)
    return sep.join(strng[:pos]), sep.join(strng[pos:])
```

```python
os.chdir(".\\impf")
files=os.listdir()
os.chdir(r"..")
dates=[pd.to_datetime(split(file,".",3)[0],format='%d.%m.%Y') for file in files]
print(dates)
```

```python
dfs_list=[pd.read_excel(f".\\impf\\{file}",sheet_name=1) for file in files]
dfs=pd.Series(dfs_list,index=dates)
dfs=dfs.sort_index()
dfs[0]
```

```python
pd.merge(
    left=shape_df, 
    right=dfs[0], 
    on="Bundesland",
    how="inner"
)
```
