```python
import math
import numpy as np
from matplotlib import pyplot as plt

#GO 1.1
delta=12.5
dgb=dg=db=0.5/1000
dg=0.5/1000+0.01

g=(np.array([33.5,37.3,40.15])-delta)/100
b=((np.array([53,55,57])-delta)/100)-g


def gauss(d):
    return math.sqrt((1/len(d)**2)*sum(d**2))
    
def f1(g,b):
    return g*b/(g+b)

def df1(g,b,f):
    return f*math.sqrt((b*dg/(g*(b+g)))**2+(g*db/(b*(b+g)))**2)

                        
def f1avg(g,b):
                        
    f=np.array([f1(x,y) for x,y in zip(g,b)])
    df=np.array([df1(x,y,z) for x,y,z in zip(g,b,f)])
    dfmean=gauss(df)
    return np.mean(f),f,dfmean,df
                
print("Brennweiten GO1.1         " + str(f1avg(g,b)[1]))
print("Brennweitenfehler GO1.1   " + str(f1avg(g,b)[3]))
print("Brennweite AVG GO1.1      " + str(f1avg(g,b)[0]))
print("BrennweitefehlerAVG GO1.1 " + str(f1avg(g,b)[2]))
print("--------------")

#GO 1.2
nadelta=11.5
#Natrium
ge=(np.array([60,65,70,75,80])-nadelta)/100
linse1_g=(np.array([25.1,24,23.55,23.15,22.7])-nadelta)/100
linse2_g=(np.array([44.6,50.1,55.5,60.9,66.3])-nadelta)/100
gd=linse2_g-linse1_g

print(ge)
print(gd)

de=0.5/1000
dd=math.sqrt(dg**2+db**2)


def f2(e,d):
    return 0.25*(e-d**2/e)

def df2(e,d):
    return math.sqrt((1/4*((d/e)**2+1)*de)**2+(-d/(2*e)*dd)**2)

def f2avg(e,d):
    f=np.array([f2(x,y) for x,y in zip(e,d)])
    df=np.array([df2(x,y) for x,y in zip(e,d)])
    dfmean=gauss(df)
    return np.mean(f),f,dfmean,df
    
print("Brennweiten GO1.2 Natrium          " + str(f2avg(ge,gd)[1]))
print("Brennweitenfehler GO1.2 Natrium    " + str(f2avg(ge,gd)[3]))
print("Brennweite AVG GO1.2 Natrium       " + str(f2avg(ge,gd)[0]))
print("Brennweitefehler AVG GO1.2 Natrium " + str(f2avg(ge,gd)[2]))
print("--------------")

#GO 1.3

chrdelta=12.5#naja 18.8 vllt
#weiß opt.
we=schirm_w=(np.array([60,65,70,75,80])-delta)/100
linse1_w=(np.array([26.45,25.45,24.75,24.3,24.2])-delta)/100
linse2_w=(np.array([44.00,50.25,55.85,61.05,66.30])-delta)/100
wd=linse2_w-linse1_w


#blau
be=schirm_w=(np.array([55,60,65,70,75])-chrdelta)/100
linse1_b=(np.array([27.6,25.9,25.3,24.6,24.3])-chrdelta)/100
linse2_b=(np.array([39,45,50.7,56,61.35])-chrdelta)/100
bd=linse2_b-linse1_b

#rot
re=schirm_w=(np.array([55,60,65,70,75])-chrdelta)/100
linse1_r=(np.array([29.2,26.4,25.6,25.1,24.35])-chrdelta)/100
linse2_r=(np.array([36.64,44.45,50.4,55.5,61.25])-chrdelta)/100
rd=linse2_r-linse1_r

print(we)
print(wd)
print("Brennweiten GO1.3 Weiß          " + str(f2avg(we,wd)[1]))
print("Brennweitenfehler GO1.3 Weiß    " + str(f2avg(we,wd)[3]))
print("Brennweite AVG GO1.3 Weiß       " + str(f2avg(we,wd)[0]))
print("Brennweitefehler AVG GO1.3 Weiß " + str(f2avg(we,wd)[2]))
print("--------------")

print(re)
print(rd)
print("Brennweiten GO1.3 Rot          " + str(f2avg(re,rd)[1]))
print("Brennweitenfehler GO1.3 Rot    " + str(f2avg(re,rd)[3]))
print("Brennweite AVG GO1.3 Rot       " + str(f2avg(re,rd)[0]))
print("Brennweitefehler AVG GO1.3 Rot " + str(f2avg(re,rd)[2]))
print("--------------")


print(be)
print(bd)
print("Brennweiten GO1.3 Blau          " + str(f2avg(be,bd)[1]))
print("Brennweitenfehler GO1.3 Blau    " + str(f2avg(be,bd)[3]))
print("Brennweite AVG GO1.3 Blau       " + str(f2avg(be,bd)[0]))
print("Brennweitefehler AVG GO1.3 Blau " + str(f2avg(be,bd)[2]))
print("--------------")



#GO1.4
#mittelstrahlen
gm=(np.array([79.9,90.8,101])-delta)/100
bm=((np.array([100,110,120])-delta)/100)-gm
#randstrahlen
gr=(np.array([81.35,91.5,102.25])-delta)/100
br=((np.array([100,110,120])-delta)/100)-gr


print(gm)
print(bm)
print("Brennweiten GO1.4 Mittelstrahlen         " + str(f1avg(gm,bm)[1]))
print("Brennweitenfehler GO1.4 Mittelstrahlen   " + str(f1avg(gm,bm)[3]))
print("Brennweite AVG GO1.4 Mittelstrahlen      " + str(f1avg(gm,bm)[0]))
print("BrennweitefehlerAVG GO1.4 Mittelstrahlen " + str(f1avg(gm,bm)[2]))
print("--------------")

print(gr)
print(br)
print("Brennweiten GO1.4 Randstrahlen         " + str(f1avg(gr,br)[1]))
print("Brennweitenfehler GO1.4 Randstrahlen   " + str(f1avg(gr,br)[3]))
print("Brennweite AVG GO1.4 Randstrahlen      " + str(f1avg(gr,br)[0]))
print("BrennweitefehlerAVG GO1.4 Randstrahlen " + str(f1avg(gr,br)[2]))
print("--------------")
```
